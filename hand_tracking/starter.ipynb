```python
import cv2
import mediapipe as mp
import glob
import pandas as pd
mp_drawing = mp.solutions.drawing_utils
mp_hands = mp.solutions.hands

file_list =glob.glob("./datasets/paper/*")
landmarks_list = []
with mp_hands.Hands(
    static_image_mode=True,
    max_num_hands=2,
    min_detection_confidence=0.5) as hands:
  for file in file_list:
    # Read an image, flip it around y-axis for correct handedness output (see
    # above).
    image = cv2.flip(cv2.imread(file), 1)
    # Convert the BGR image to RGB before processing.
    results = hands.process(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    if results.multi_hand_landmarks:
        landmarks_list.append(results.multi_hand_landmarks[0])
```

```python
x_lst = []
y_lst = []
z_lst = []
for l in range(len(landmarks_list)):
    lst = list(str(landmarks_list[l]).split(sep=None, maxsplit=-1))
    x = []
    y = []
    z = []

    for i in range(len(lst)):
        if 'x' in str(lst[i]):
            x.append(lst[i+1])
        elif 'y' in str(lst[i]):
            y.append(lst[i+1])
        elif 'z' in str(lst[i]):
            z.append(lst[i+1])

    x_lst.append(x)
    y_lst.append(y)
    z_lst.append(z)
```

```python
df = pd.DataFrame()
for i in range(21):
    column_name = str(f'ID {i}')
    df[column_name] = 0
```

```python
for i in range(len(landmarks_list)):
    row_lst = []
    for c in range(21):
        row_lst.append([x_lst[i][c], y_lst[i][c]])
    df.loc[i] = row_lst
```
